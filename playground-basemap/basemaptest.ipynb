```python
## This code taken from: https://waterprogramming.wordpress.com/2017/04/03/making-watershed-maps-in-python/
## Modified for other purposes by UMA

from mpl_toolkits.basemap import Basemap
from matplotlib import pyplot as plt
 
fig = plt.figure()
fig.set_size_inches([17.05,8.15])
ax = fig.add_subplot(111)

# m = Basemap(projection = 'merc', llcrnrlat=10, urcrnrlat=50,
#        llcrnrlon=-160, urcrnrlon=-60)

# plot basemap, rivers and countries
m = Basemap(llcrnrlat=-7.25, urcrnrlat=-7.05, llcrnrlon=107.55, urcrnrlon=107.75, resolution='h')
m.arcgisimage(service='World_Shaded_Relief')
m.drawrivers(color='dodgerblue',linewidth=1.0,zorder=1)
m.drawcountries(color='k',linewidth=1.25);
```

```python
import numpy as np
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection
 
# plot Red River basin
m.readshapefile('./shp/DAS_uma', 'Basin', drawbounds=False)
patches = []
for info, shape in zip(m.Basin_info, m.Basin):
    #if info['DAS Paceut'] == 1: # attribute in attribute table of shapefile
    patches.append(Polygon(np.array(shape), True))
 
ax.add_collection(PatchCollection(patches, facecolor='0.33', edgecolor='0.5', alpha=0.5));
```
