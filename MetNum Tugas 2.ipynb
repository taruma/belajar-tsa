# Metode Numerik

## FTCS
$$ h_{i}^{n+1} = h_{i}^{n} + \frac{\Delta t}{\Delta x^{2}} \frac{T}{S} \left( h_{i+1}^{n}-2 h_{i}^{n}+h_{i-1}^{n} \right)$$

```python
import numpy as np
import matplotlib.pyplot as plt

h1 = 16
h2 = 11
T = 0.02
S = 0.002

L = 100
dx = 20

imax = int(L/dx) + 1

dt = 10
t = 100
tmax = int(t/dt)

h = np.zeros(imax)
hn = np.zeros(imax)
absis = np.linspace(0,100,6)

# Initial
for i in range(0, imax):
    h[i] = h1

for i in range(0, tmax):

    for j in range(1, imax-1):
        hn[j] = h[j] + (dt/(dx**2)) * (T/S) * (h[j+1]-2*h[j]+h[j-1])
    
    # Syarat batas
    hn[0] = h1
    hn[5] = h2
    
    h = np.copy(hn)
    plt.plot(absis, h, 'b--');
    print('Iterasi ke - ',i , "=== ", h,'\n')
```

## Dufort-Frankel Scheme

$$ 1 + \frac{}{} $$

```python
import numpy as np
import matplotlib.pyplot as plt

h1 = 16
h2 = 11
T = 0.02
S = 0.002

L = 100
dx = 20

imax = int(L/dx) + 1

dt = 10
t = 100
tmax = int(t/dt)

h0 = np.zeros(imax)
h = np.zeros(imax)
hn = np.zeros(imax)
absis = np.linspace(0,L,imax)

# Initial
for i in range(0, imax):
    h[i] = h1
    h0[i] = h1
    
# Simulasi
LA = 1 + (T/S) * (2*dt/(dx**2))
RA = 1 - (T/S) * (2*dt/(dx**2))
RB = (T/S) * (2*dt/(dx**2))


for i in range(0, tmax):

    for j in range(1, imax-1):
        hn[j] = (RA*h0[j] + RB*(h[j+1] - h[j-1]))/LA
    
    # Syarat batas
    hn[0] = h1
    hn[5] = h2

    h0 = np.copy(h)
    h = np.copy(hn)

    print('Iterasi ke - ',i , "=== ", h,'\n')
    plt.plot(absis, h, 'r--');
```
