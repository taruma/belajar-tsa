## FTCS
$$ h_{i}^{n+1} = h_{i}^{n} + \frac{\Delta t}{\Delta x^{2}} \frac{T}{S} \left( h_{i+1}^{n}-2 h_{i}^{n}+h_{i-1}^{n} \right)$$ 

```python
import numpy as np
import matplotlib.pyplot as plt

h1 = 16
h2 = 11
T = 0.02
S = 0.002

L = 100
dx = 20

imax = int(L/dx) + 1

dt = 10
t = 100
tmax = int(t/dt)

h = np.zeros(imax)
hn = np.zeros(imax)
absis = np.linspace(0, 100, 6)

waktu = 0

# Initial
for i in range(0, imax):
    h[i] = h1

for i in range(0, tmax):

    for j in range(1, imax-1):
        hn[j] = h[j] + (dt/(dx**2)) * (T/S) * (h[j+1]-2*h[j]+h[j-1])

    # Syarat batas
    hn[0] = h1
    hn[5] = h2
    
    waktu += dt
    h = np.copy(hn)
    plt.plot(absis, h, 'b--')
    print('Waktu ke - ', waktu, "=== \t", h, '\n')

plt.xlabel('jarak (m)')
plt.ylabel('muka air (m)')
msg = 'dt = ' + str(dt) + ' dx= ' + str(dx) + ' Time = ' + str(t)
plt.title('Elevasi muka air - FTCS ' + msg);
```

## Dufort-Frankel Scheme

```python
import numpy as np
import matplotlib.pyplot as plt

h1 = 16
h2 = 11
T = 0.02
S = 0.002

L = 100
dx = 20

imax = int(L/dx) + 1

dt = 10
t = 100
tmax = int(t/dt)

h0 = np.zeros(imax)
h = np.zeros(imax)
hn = np.zeros(imax)
absis = np.linspace(0, L, imax)

# Initial
for i in range(0, imax):
    h[i] = h1
    h0[i] = h1

# Simulasi
LA = 1 + (T/S) * (2*dt/(dx**2))
RA = 1 - (T/S) * (2*dt/(dx**2))
RB = (T/S) * (2*dt/(dx**2))

waktu = 0



for i in range(0, tmax):

    for j in range(1, imax-1):
        hn[j] = (RA*h0[j] + RB*(h[j+1] + h[j-1]))/LA

    # Syarat batas
    hn[0] = h1
    hn[5] = h2

    h0 = np.copy(h)
    h = np.copy(hn)

    waktu += dt
    print('Waktu ke - ', waktu, "=== \t", h, '\n')
    plt.plot(absis, h, 'r--')

plt.xlabel('jarak (m)')
plt.ylabel('muka air (m)')
msg = 'dt = ' + str(dt) + ' dx= ' + str(dx) + ' Time = ' + str(t)
plt.title('Elevasi muka air - Dufort Frankel ' + msg);
```

## Leasonen

```python
import numpy as np
import matplotlib.pyplot as plt

h1 = 16
h2 = 11
T = 0.02
S = 0.002

L = 100
dx = 20

imax = int(L/dx) + 1

dt = 10
t = 100
tmax = int(t/dt)

h = np.zeros(imax)
hn = np.zeros(imax)
absis = np.linspace(0, L, imax)

# Initial
for i in range(0, imax):
    h[i] = h1

hn[0] = h1
hn[5] = h2
    
# Simulasi

PA = T/S * (dt/dx**2)
PB = 1 + T/S * (2*dt/dx**2)
PC = T/S * (dt/dx**2)

waktu = 0

for i in range(0, tmax):

    for j in range(1, imax-1):
        hn[j] = (h[j] + PA*(hn[j+1]+hn[j-1]))/PB
        for k in range(1, j):
            hn[j-k] = (h[j-k] + PA*(hn[j-k+1]+hn[j-k-1]))/PB

    # Syarat batas
    hn[0] = h1
    hn[5] = h2

    h = np.copy(hn)
    
    waktu += dt
    print('Waktu ke - ', waktu, "=== \t", h, '\n')
    plt.plot(absis, h, 'k--')

plt.xlabel('jarak (m)')
plt.ylabel('muka air (m)')
msg = 'dt = ' + str(dt) + ' dx= ' + str(dx) + ' Time = ' + str(t)
plt.title('Elevasi muka air - Leasonen ' + msg);
```

## Crank-Nicholson

```python
import numpy as np
import matplotlib.pyplot as plt

h1 = 16
h2 = 11
T = 0.02
S = 0.002

L = 100
dx = 20

imax = int(L/dx) + 1

dt = 10
t = 100
tmax = int(t/dt)

h = np.zeros(imax)
hn = np.zeros(imax)
absis = np.linspace(0, L, imax)

# Initial
for i in range(0, imax):
    h[i] = h1

hn[0] = h1
hn[5] = h2

# Simulasi

CA = - T/S * (dt/(2*dx**2))
CB = 1 + T/S * (dt/dx**2)
CC = CA
CD = 0

waktu = 0

for i in range(0, tmax):

    for j in range(1, imax-1):
        CD = h[j] + T/S*dt/(2*dx**2) * (h[j-1]-2*h[j] + h[j+1])
        hn[j] = (CD - CA * (hn[j-1] + hn[j+1]))/CB
        for k in range(1, j):
            CD = h[j-k] + T/S*dt/(2*dx**2) * (h[j-k-1] - 2*h[j-k] + h[j-k+1])
            hn[j-k] = (CD - CA * (hn[j-k-1] + hn[j-k+1]))/CB

    # Syarat batas
    hn[0] = h1
    hn[5] = h2

    h = np.copy(hn)

    waktu += dt
    print('Waktu ke - ', waktu, "=== \t", h, '\n')
    plt.plot(absis, h, 'm--')

plt.xlabel('jarak (m)')
plt.ylabel('muka air (m)')
msg = 'dt = ' + str(dt) + ' dx= ' + str(dx) + ' Time = ' + str(t)
plt.title('Elevasi muka air - Crank-Nicholson ' + msg);
```
