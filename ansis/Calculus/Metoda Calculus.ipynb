## Calculus Single Variable

```python
from sympy import symbols, latex, simplify, init_printing, solve, solveset, Eq
from sympy import Rational as rat
init_printing()

sx = symbols('x')

fx = (1/4)*sx**4 - (1/2)*sx**2
fxd = fx.diff(sx)
fxdd = fx.diff(sx,sx)

def f(x):
    return float(fx.subs({sx: x}))

def fd(x):
    return float(fxd.subs({sx: x}))

def fdd(x):
    return float(fxdd.subs({sx: x}))
```

```python
import numpy as np
import matplotlib.pyplot as plt

axis = np.linspace(xa,xb,1000)
y, yd, ydd = [], [], []
for x in axis:
    y.append(f(x))
    yd.append(fd(x))
    ydd.append(fdd(x))

plt.rcParams["figure.figsize"] = (10, 12)
plt.plot(axis, y)
plt.plot(axis, yd)
plt.plot(axis, ydd)
plt.grid()
```
